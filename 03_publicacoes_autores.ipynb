## Etapa 4 – Verificação de dados ausentes/duplicados & tratamento

```python
missing_percentage_per_column = df.isnull().mean() * 100
missing_percentage_per_column = missing_percentage_per_column.sort_values(ascending=False)
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x=missing_percentage_per_column.index, y=missing_percentage_per_column.values)
plt.xticks(rotation=45)
plt.xlabel('Colunas')
plt.ylabel('Percentual de Dados Ausentes')
plt.title('Percentual de Dados Ausentes por Coluna')
plt.show()
```

## Etapa 5 – Análises 

```python
lista_orgao = df['Título'].value_counts().reset_index()
lista_orgao.columns = ['Título', 'contagem']
lista_orgao

#Análise - O mesmo título aparece mais de 1 vez, o que pode significar que existam mais de um coautor na mesma base. 
```

```python
lista_AUTORES = df.groupby('Nome').agg({
    'Título': [
        ('contagem', 'count'),
        ('publicações', lambda x: ', '.join(sorted(set(x.dropna()))))
    ]
}).reset_index()

# Ajustar o nome das colunas (multi-index resultante do groupby com múltiplas agregações)
lista_AUTORES.columns = ['Nome', 'Contagem', 'Publicações']
lista_AUTORES = lista_AUTORES.sort_values(by='Contagem', ascending=False)

lista_AUTORES.head(10)
```
